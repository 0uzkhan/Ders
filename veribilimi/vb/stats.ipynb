```python
#medyan hesaplama ?

sorted_salaries = sorted(salaries)  
adet = len(sorted_salaries)

if adet % 2 == 1:  
    medyan = sorted_salaries[adet // 2]
else:  
    medyan = (sorted_salaries[adet // 2 - 1] + sorted_salaries[adet // 2]) / 2

print("Medyan:", medyan)
```

```python
def medyan_hesaplama(series):
    sorted_series = sorted(series)
    adet = len(sorted_series)

    if adet % 2 == 1:
        return sorted_series[adet // 2]
    else:
        return (sorted_series[adet // 2 - 1] + sorted_series[adet // 2]) / 2

print("Medyan:", medyan_hesaplama(salaries))
```

```python
#kartiller ? Q1,Q2,Q3,Q4

import numpy as np

sorted_salaries = sorted(salaries)

Q1 = np.percentile(sorted_salaries, 25)  
Q2 = np.percentile(sorted_salaries, 50)  
Q3 = np.percentile(sorted_salaries, 75) 
Q4 = max(sorted_salaries)  

print("Q1 (Alt çeyrek):", Q1)
print("Q2 (Medyan):", Q2)
print("Q3 (Üst çeyrek):", Q3)
print("Q4 (Maksimum Değer):", Q4)
```

```python
def kartil_hesaplama(series):
    sorted_series = sorted(series)
    Q1 = np.percentile(sorted_series, 25)
    Q2 = np.percentile(sorted_series, 50)
    Q3 = np.percentile(sorted_series, 75)
    Q4 = max(sorted_series)
    
    return Q1, Q2, Q3, Q4

q1, q2, q3, q4 = kartil_hesaplama(salaries)
print(f"Q1: {q1}, Q2: {q2}, Q3: {q3}, Q4: {q4}")
```

```python
#standart sapma? 

ortalama = sum(salaries) / len(salaries)

toplam_kare_fark = sum((i - ortalama) ** 2 for i in salaries)
standart_sapma = (toplam_kare_fark / len(salaries)) ** 0.5  # Popülasyon standart sapması

print("Standart Sapma:", standart_sapma)
```

```python
def standart_sapma_hesaplama(series):
    ortalama = sum(series) / len(series)
    toplam_kare_fark = sum((i - ortalama) ** 2 for i in series)
    return (toplam_kare_fark / len(series)) ** 0.5

print("Standart Sapma:", standart_sapma_hesaplama(salaries))
```

```python
#varyans?

ortalama = sum(salaries) / len(salaries)

toplam_kare_fark = sum((i - ortalama) ** 2 for i in salaries)
varyans = toplam_kare_fark / len(salaries)  # Popülasyon varyansı

print("Varyans:", varyans)
```

```python
def varyans_hesaplama(series):
    ortalama = sum(series) / len(series)
    toplam_kare_fark = sum((i - ortalama) ** 2 for i in series)
    return toplam_kare_fark / len(series)  

print("Fonksiyon ile Varyans:", varyans_hesaplama(salaries))
```

```python
#çarpıklık

import numpy as np
from scipy.stats import skew

iqs = [44,45,46,47,47,48,48,49,50,50,50,50,51,51,51,51,51,51]

ortalama = np.mean(iqs)
standart_sapma = np.std(iqs, ddof=1)  # ddof=1 çünkü örneklem standart sapması

n = len(iqs)
skewness = (n / ((n-1) * (n-2))) * sum(((x - ortalama) / standart_sapma) ** 3 for x in iqs)

print("Çarpıklık:", skewness)
```

```python
import seaborn as sns
sns.distplot(iqs)
```

```python
#iqs verisinde ortalama hesaplama
#median
#iqs verisinde standart sapma
# (3 * (mean_iq - median_iq)) / std_iq çarpıklık katsayısı formulu
```

Çarpıklık katsayımız 0 ise dağılım simetriktir, normal dağılımdır.

Çarpıklık katsayımız küçüktür 0'dan ise dağılım negatif çarpıktır.

Çarpıklık katsayımız büyüktür 0'dan ise dağılım pozitif çarpıktır.

```python
mean_iq = np.mean(iqs)
print("Ortalama (Mean):", mean_iq)
```

```python
median_iq = np.median(iqs)
print("Medyan (Median):", median_iq)
```

```python
std_iq = np.std(iqs, ddof=1)  # Örneklem standart sapması için ddof=1 kullanılır.
print("Standart Sapma (Standard Deviation):", std_iq)
```

```python
skewness_iq = (3 * (mean_iq - median_iq)) / std_iq
print("Çarpıklık Katsayısı (Skewness):", skewness_iq)
```

```python
sizes = [40,42,42,44,44,44,46,46,48]

sns.distplot(sizes,bins=5)
```

```python
#çarpıklık katsayısı hesaplama
from scipy.stats import skew

skewness_scipy = skew(sizes, bias=False)  # bias=False örneklem çarpıklığını hesaplar
print("Çarpıklık Katsayısı (SciPy ile):", skewness_scipy)
```

```python
#kurtosis --- basıklık katsayısı, iqs verisinde dene, sizes verisinde dene, distplot
```

```python
from scipy.stats import kurtosis

kurtosis_iq = kurtosis(iqs, fisher=True)  # Fisher=True → Normal dağılıma göre hesaplar
kurtosis_sizes = kurtosis(sizes, fisher=True)

print("iqs verisi için Kurtosis (Basıklık Katsayısı):", kurtosis_iq)
print("sizes verisi için Kurtosis (Basıklık Katsayısı):", kurtosis_sizes)
```
